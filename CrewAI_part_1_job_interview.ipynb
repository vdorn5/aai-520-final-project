```python
### Let's do a small implementation of each model with a question using crewai
from crewai import LLM

# Let's create a small query
query = "Can you list the ingredients for bandeja paisa?"

llm_gptoss = LLM(
    model = "ollama/gpt-oss:20b",
    base_url="http://localhost:11434")

res = llm_gptoss.call(query)
print(res)
```

# Function for different llms
We are gonna create a function called `get_llm` with specific arguments in order to retrieve a model using `ChatOpenAI` or `ollama` hosted models.

```python
from typing import Union, Any

def get_llm(model_name:str, hosting_type:str = "ollama") -> Union[str, LLM, Any]:
  if hosting_type == "ollama":
    return LLM(
        model = f"ollama/{model_name}",
        base_url="http://localhost:11434",
        temperature = 0.5)
  else:
    return ChatOpenAI(
        model_name=model_name,
        temperature = 0.5)
```

# Get the data

* Interviewer.
* company.
* job position.
* job description.

```python
# Store the data
interviewer = input("Interviewer: ")
company = input("Company: ")
job_position = input("Job position: ")
job_description = input("Job description: ")
```

```python
# Backstory
interviewer_backstory = input("Interviewer's backstory: ")
```

```python
### Let's get the llm
possible_names = ["gpt-4o", "gpt-oss:20b", "gemma3", "qwen3:8b"]
#llm = get_llm(model_name = possible_names[0], hosting_type = "other")
llm = get_llm(model_name = possible_names[2], hosting_type = "ollama")
```

### Interviewer AI agent


```python
# Create the interviewer agent
interviewer_agent = Agent(
    role = f"You are {interviewer}, employed at {company}, and hiring for job position: {job_position}",
    goal = f"You help the user prepare for the job interview for job position: {job_position} with the following description: {job_description}",
    backstory= interviewer_backstory,
    llm = llm)
```

```python
# Define the interviewer prep_task
interviewer_prep_task = Task(
    description = f"""
    Interview the user for the job {job_position} with the following description: {job_description}
    """,
    expected_output=f"""
    Ask only one question for the user that is relevant for the job {job_position}.
    """,
    agent=interviewer_agent,
    human_input = True
)
```

# AI Coach

```python
# Build the AI interview coach agent
coach_agent = Agent(
    role = "Interview Coach",
    goal = f"Help the user prepare for the job interview for job position: {job_position} by grading the relevance of the answer",
    backstory= "You are an expert in job interviews",
    llm = llm)
```

```python
# Define the coaching Task
coaching_task = Task(
    description = f""" Give feedback to the user on its answer """,
    expected_output= f""" Give feedback to the user on its answer with bullet points on what was good, bad, and how to take it to the next level""",
    agent=coach_agent,
    context = [interviewer_prep_task]
)
```

```python
crew = Crew(
    agents = [interviewer_agent, coach_agent],
    tasks = [interviewer_prep_task, coaching_task],
    verbose = True,
    process = Process.sequential,
)
result = crew.kickoff()

#Store each question
# interview_questions.append(result.tasks_output[0].raw)
# interview_content = "".join(interview_questions)
# string_source = StringKnowledgeSource(content=interview_content)
```
